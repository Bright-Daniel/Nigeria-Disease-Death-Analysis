```python
# Let me start by importing the required libraries
import pandas as pd
import numpy as np

# importation of dataset
death_d = pd.read_csv('cause_of_deaths.csv')
death_d.sample(30)
```

**_The data required for this analysis which I am going to extract with few blocks of code from death_d are rows containing Nigeria because I will be considering causes of data in Nigeria only_**

```python
# Extraction of the rows containing Nigeria
new_t= death_d[death_d['Code'].str.contains('NGA')]
new_t.sample(10)
```
